```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
data = pd.read_csv('/content/data.csv')
```

```python
item_sequences = data.groupby('category')['title'].apply(list).tolist()
print("Item sequences by category:")
print(item_sequences[:5])
```

```python
model = Word2Vec(sentences=item_sequences, vector_size=50, window=5, min_count=1, workers=4)
```

```python
item_vectors = {item: model.wv[item] for item in model.wv.index_to_key}
item_titles = list(item_vectors.keys())
embedding_matrix = np.array([item_vectors[item] for item in item_titles])
```

```python
similarity_matrix = cosine_similarity(embedding_matrix)
```

```python
title_to_index = {title: idx for idx, title in enumerate(item_titles)}
index_to_title = {idx: title for title, idx in title_to_index.items()}
```

```python
def get_recommendations(item_title, similarity_matrix, top_n=5):
    if item_title not in title_to_index:
        print(f"Item title '{item_title}' not found in the model.")
        return []
    item_idx = title_to_index[item_title]
    similar_indices = np.argsort(similarity_matrix[item_idx])[::-1][1:top_n+1]
    return [index_to_title[i] for i in similar_indices]
```

```python
item_title = 'Example Product Title'  # Replace with an actual title from your data
recommendations = get_recommendations(item_title, similarity_matrix, top_n=5)
print(f"Recommendations for '{item_title}':", recommendations)
```

```python
from sklearn.metrics import precision_score, recall_score
import numpy as np
```

```python
def precision_at_k(recommended_items, relevant_items, k):
    recommended_at_k = recommended_items[:k]
    relevant_at_k = [1 if item in relevant_items else 0 for item in recommended_at_k]
    return sum(relevant_at_k) / k
```

```python

def recall_at_k(recommended_items, relevant_items, k):
    recommended_at_k = recommended_items[:k]
    relevant_at_k = [1 if item in relevant_items else 0 for item in recommended_at_k]
    return sum(relevant_at_k) / len(relevant_items)
```

```python
def average_precision(recommended_items, relevant_items):
    num_relevant = 0
    precisions = []

    for i, item in enumerate(recommended_items):
        if item in relevant_items:
            num_relevant += 1
            precisions.append(num_relevant / (i + 1))

    if not precisions:
        return 0.0

    return sum(precisions) / len(relevant_items)
```

```python
def mean_average_precision(recommendations, ground_truth):
    aps = []

    for user_id in recommendations:
        recommended_items = recommendations[user_id]
        relevant_items = ground_truth.get(user_id, [])
        ap = average_precision(recommended_items, relevant_items)
        aps.append(ap)

    return np.mean(aps)
```

```python
def ndcg_at_k(recommended_items, relevant_items, k):
    def dcg(scores):
        return sum([score / np.log2(idx + 2) for idx, score in enumerate(scores)])

    recommended_at_k = recommended_items[:k]
    relevance_scores = [1 if item in relevant_items else 0 for item in recommended_at_k]
    dcg_score = dcg(relevance_scores)

    ideal_relevance_scores = sorted(relevance_scores, reverse=True)
    idcg_score = dcg(ideal_relevance_scores)

    return dcg_score / idcg_score if idcg_score > 0 else 0.0
```

```python
recommendations = {
    'user_1': ['item_5', 'item_12', 'item_123'],
    'user_2': ['item_7', 'item_42', 'item_23']
}

ground_truth = {
    'user_1': ['item_5', 'item_123', 'item_200'],
    'user_2': ['item_7', 'item_42']
}

k = 3
map_score = mean_average_precision(recommendations, ground_truth)
print(f"Mean Average Precision (MAP): {map_score:.4f}")

for user_id, recommended_items in recommendations.items():
    relevant_items = ground_truth.get(user_id, [])
    precision = precision_at_k(recommended_items, relevant_items, k)
    recall = recall_at_k(recommended_items, relevant_items, k)
    ndcg = ndcg_at_k(recommended_items, relevant_items, k)
    print(f"User {user_id} - Precision@{k}: {precision:.4f}, Recall@{k}: {recall:.4f}, NDCG@{k}: {ndcg:.4f}")
```
